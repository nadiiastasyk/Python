1. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{(y_{i} - (b + mx_{i}))^{2}}$$

&emsp;&emsp;де x,y - деякі задані вектори

```python
import numpy as np

np.random.seed(42)
x = np.random.normal(0, 5, 100)
y = 50 + 2 * x + np.random.normal(0, 2, len(x))


b = np.ones_like(x)
m = np.ones_like(y)

# Визначимо функцію та її похідну:
def f(b, m):
    return np.mean(np.power((y - (b + m * x)), 2))

def df(b, m):
    h = 1e-10
    f_b = (f(b + h, m) - f(b, m)) / h
    f_m = (f(b, m + h) - f(b, m)) / h
    return np.array([f_b, f_m])

# шукаємо мінімальне значення
w0 = np.ones(2)  # значення w, з якого ми почнемо пошук
alpha = 0.01  # крок алгоритма
epsilon = 0.00001  # критерій, якщо зміна w менше "epsilon", то вважаємо, що рішення знайдено

n = 0
while True:
    w = w0 - alpha * df(b, m)  # крок алгоритму, перехід до наступного значення w
    
    if np.sum(np.abs(w - w0)) <= epsilon:  # перевірка (тут, можливо, пошук варто припинити)
        break
    else:
        w0 = w  # оновлення попереднього значення

    n += 1
    if n >= 100:
        break

    print(w, f(w[0], w[1]))  # виведення поточного значення х та y

print('Значення w, при якому f набуває мінімального значення: ', f(w[0], w[1]))
print('Мінімум f w: ', w)
```

2. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{[(y_{i} - (b + mx_{i}))^{2}]} + m^2 + b^2$$

&emsp;&emsp;де x,y - вектори з 1-го завдання

3. Знайти значення а, для якого функція набуває мінімального значення

$$f = -\frac{1}{n}\sum_{i=1}^{n}{[y_{i}log(h(x_{i})) + (1 - y_{i})log(1 - h(x_{i}))]}$$

$$h = \frac{1} {1 + exp^{-ax}}$$

```python
np.random.seed(10)
num_observations = 100
x1 = np.random.normal(1, 1, 50)
x2 = np.random.normal(3, 1, 50)

x = np.concatenate([x1, x2])
y = np.concatenate([np.zeros_like(x1), np.ones_like(x2)])

a = np.ones_like(x)
```

```python
# Визначимо функцію та її похідну:
def f(x, y, a):
    h = 1/(1 + np.exp(-a * x))
    return np.mean(y * np.log(h) + (1 - y) * np.log(1 - h))

def df(x, y, a):
    h = 1e-10
    f_x = (f(x + h, y, a) - f(x, y, a)) / h
    f_y = (f(x, y + h, a) - f(x, y, a)) / h
    return np.array([f_x, f_y])

w0 = np.ones(2)  # значення w, з якого ми почнемо пошук
alpha = 0.01  # крок алгоритма
epsilon = 0.00001  # критерій, якщо зміна w менше "epsilon", то вважаємо, що рішення знайдено

print(w0, f(x, y, a))

n = 0
while True:
    
    w = w0 - alpha * df(x, y, a)  # крок алгоритму, перехід до наступного значення w
    
    if np.sum(np.abs(w - w0)) <= epsilon:  # перевірка (тут, можливо, пошук варто припинити)
        break
    else:
        w0 = w  # оновлення попереднього значення
    
    n += 1
    if n >= 100:
        break
        
    print(w, f(w[0], w[1], a)) 

print('Значення w, при якому f набуває мінімального значення: ', f(w[0], w[1], a))
print('Мінімум f: ', w)
```
