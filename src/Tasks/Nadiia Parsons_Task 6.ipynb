1. Створити графіки для віку пасажирів та ціни білетів на одній області та на двох областях (одній над іншою).

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

titanic_data = pd.read_csv(
    '../Lessons/Lesson 5/Files/TitanicDataset.csv',
    sep=',',
    header=0)

fig = plt.figure()
ax1 = fig.add_axes([0.125, 0.547727, 0.775, 0.352273])
ax1.set_title(u'Область 1')

ax1.scatter(titanic_data['PassengerId'], titanic_data['Fare'], color='green')
ax2 = fig.add_axes([0.125, 0.125, 0.775, 0.352273])
ax2.set_title(u'Область 2')
ax2.scatter(titanic_data['PassengerId'], titanic_data['Age'], color='red')
    
plt.show()
```

2. Створити гістограму для статі пасажирів (кількість пасажирів тої чи іншої статі) та кількості що вижили залежно від статі.

```python
fig2 = plt.figure()
titanic_data.groupby(by=['Sex']).agg(
    sex=('Sex', 'count'),
    survived=('Survived', 'sum')
).plot.bar()

plt.show()
```

3. Створити гістограму для класу квитка пасажирів (кількість пасажирів відповідного класу) та кількості що вижили залежно від статі.

```python
fig3 = plt.figure()
# створюємо 2 графіки
fig3, (ax1, ax2) = plt.subplots(1, 2)

# будуємо першу гістограму
ax1.hist(titanic_data['Pclass'], histtype ='bar')
ax1.set_title('Пасажири відповідно класу')
ax1.set_xlabel('Клас')
ax1.set_ylabel('Кількість пасажирів')

# будуємо другу гістограму
survived = titanic_data.groupby(by=['Sex'])['Survived'].count()
ax2.hist(survived, histtype ='bar')
ax2.set_title('Пасажири, що вижили залежно від статі')
ax2.set_xlabel('Кількість пасажирів')
ax2.set_ylabel('Стать')
```

4. Створити кругову діаграму по кількості пасажирів за місцем посадки.

```python
fig4 = plt.figure()

passengers = titanic_data.groupby(by='Embarked')['PassengerId'].count()
print(passengers)

# Додавання на малюнок кругової області малювання
ax = fig4.add_axes([0, 0, 1, 1])
plt.pie(passengers, labels=passengers)
plt.legend(titanic_data['Embarked'])
# plt.savefig('Polar figure.png')

plt.show()
```

5. Побудуйте графіки функцій на відповідних проміжках:\
a. y=x<sup>3</sup>-2x+1, -4≤x≤4\
b. y=$\cos$ 3x -$\sin$ x , -4 $\Pi$ ≤x≤4 $\Pi$\
c. $y= { x^2 + {1 \over x}, -4≤x≤4 }$

```python
fig5 = plt.figure()

# a plot
ax = np.arange(-4, 5)
ay = np.power(ax, 3) - 2 * ax + 1


plt.plot(ax, ay, label='$' + 'y=x^3 - 2x + 1' + '$')

# b plot
bx = np.arange(-4 * np.pi, 4 * np.pi + 1)
by = np.cos(3 * bx) - np.sin(bx)
plt.plot(bx, by, label='$' + 'y=cos(3x)-sin(x)' + '$')

# c plot
cx = np.arange(-4, 5)
cy = np.power(cx, 2) + 1/cx
plt.plot(cx, cy, label='$' + 'y=x^2 + 1/x' + '$')

plt.legend()
```
