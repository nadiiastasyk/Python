## Завдання 1

Вибрати імена пасажирів, які підпадають під такі критерії:\
     - Клас каюти 1-й чи 3-й,\
     - Вік в інтервалі від 20 до 40,\
     - Ціна квитка НЕ знаходиться в інтервалі від 70 до 100.


```python
import pandas as pd

titanic_data = pd.read_csv(
    '../Lessons/Lesson 5/Files/TitanicDataset.csv',
    sep=',',
    header=0)

filtered_data = titanic_data[
    (titanic_data['Pclass'] == 3 | 1) & 
    titanic_data['Age'].isin(range(20, 41)) &
    -titanic_data['Fare'].isin(range(70, 101))]

print(filtered_data)

filtered_data.to_csv(
    'CopyOfTitanicData.csv',  # шлях до файлу, в який робимо запис
    sep=';',  # роздільник колонок
    header=True,  # чи записувати імена колонок у файл (True - записувати, False - ні)
    index=False,  # записувати індекс у файл (True - записувати, False - ні)
)
```

## Завдання 2

```python
# считуємо дані для отримання словника
crime_data = pd.read_excel(
    '../Lessons/Lesson 5/Files/CrimesData.xlsx',
    sheet_name=None,
    header=0,  # рядок, в якому знаходяться назви стовпців
    usecols="A:F",  # імена колонок, які потрібно прочитати (None – взяти всі колонки)
    nrows=100,  # прочитати лише перші 100 рядків
    skiprows=0  # кількість рядків, які потрібно пропустити на початку файлу (читання почнеться з 0-го рядка)
)

# записуємо всі аркуші в окремий файл
with pd.ExcelWriter(
    'CrimesDataSingleDF.xlsx',
    mode='w'
) as writer:
    for sheet_name, data in crime_data.items():
        data.to_excel(
            excel_writer=writer,
            sheet_name=sheet_name,
            index=False
        )

        # додаємо кожен аркуш для отримання одного датафрейму
        data += data

print(type(data))

# правильне рішення
pd2 = pd.concat(crime_data)
pd2

```
