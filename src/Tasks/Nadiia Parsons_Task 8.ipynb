Виконати завдання за допомогою готових формул (калькуляторів) та
перевірити за допомогою відповідних розподілів (Python).

```python
from matplotlib import pyplot as plt
import numpy as np

np.random.seed(100)  # встановлюємо випадковий стан, що дорівнює довільному числу

# функція для побудови графіків щільності
def display_density(random_variable, var_name='', bins=None):
    
    # функція для побудови гістограми
    plt.hist(
        x=random_variable, # змінна, гістограму якої потрібно побудувати
        density=True, # параметр, який визначає, чи потрібно уявити гістограму як щільність розподілу
        bins=bins # у стовпців у гістограмі
    )
    
    plt.xlabel('Random variable') # помістити на графік назву горизонтальної осі
    plt.ylabel('pdf') # помістити на графік назву вертикальної осі
    plt.title(f'{var_name} density plot') # назва всього графіка
    
    plt.show() # виклик функції, яка виведе отриманий графік на екран
```

# 1. Біноміальний розподіл.
 Під час тестування з математики студент має дати правильнi відповіді на 5 запитань. Імовірність того, що вiн на позитивну оцiн- Ксу відповість на одне запитання, у середньому дорівнює 0,8. Щоб скласти тест, студентові необхiдно дати вiдповiдь не менш ніж на три питання. Знайти ймовірність того, що студент складе тест.

```python
n = 5
p = 0.8
k = 3

binomial_var = np.random.binomial(n, p, 10000)
sum([1 for item in binomial_var if item >= k]) / len(binomial_var)
```

```python
display_density(binomial_var, 'Binomial variable', bins=100)
```

# 2. Пуассонівський розподіл
На АЗС за кожну хвилину надходять у середньому два автомобілі для заправляння пальним.
 Яка ймовірність того, що за 3 хв на АЗС
для заправляння надійде: 1) один автомобіль; 2) не бiльш як три автомобiлi?


```python
poisson_var = np.random.poisson(
    lam=6, # значення параметра лямбда 2 (дзвінки за хв) *3 (хв) = 6
    size=10000 #кількість прикладів, які потрібно згенерувати
)
#1) ймовірність, що надійде один автомобіль
sum([1 for item in poisson_var if item == 1]) / len(poisson_var)
```

```python
# 2) ймовірність, що надійде не бiльш як три автомобiлi
sum([1 for item in poisson_var if item <= 3]) / len(poisson_var)
```

```python
display_density(poisson_var, 'Poisson variable', bins=100)
```

# 3.Нормальний закон розподілу.
Заряд мисливського пороху зважують на терезах, що мають
стандартне відхилення зважування 0,5 г. Середня маса порохового заряду 4,5 г. Визначити ймовірність пошкодження мисливської рушниці при пострілі, якщо максимально допустима вага порохового заряду 4,9 г. Тобто, знайдіть ймовірність того, що випадково взятий пороховий заряд буде важити більше 4,9 г.

```python
normal_var = np.random.normal(
     loc = 4.5, # Середнє значення величини
     scale=0.5, # стандартне відхилення
     size=100000 # кількість прикладів, які потрібно згенерувати
)
sum([1 for item in normal_var if item > 4.9]) / len(normal_var)
```

```python
from scipy.stats import norm
distr = norm(4.5, 0.5)
```

```python
display_density(normal_var, 'Normal variable', bins=100)
```

# 4.Рівномірний розподіл.
Випадкова величина X має рівномірний розподіл на відрізку [0; 80]. Знайти ймовірність того, що 30<X<60.

```python
uniform_var = np.random.uniform(
    low=0,
    high=81,
    size=10000)
sum([1 for item in uniform_var if item > 30 and item < 60])/len(uniform_var)
```

```python
display_density(uniform_var, 'Uniform variable', bins=100)
```
